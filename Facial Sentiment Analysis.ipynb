# Performing task on live video

```python
import cv2
from deepface import DeepFace as dep
cap=cv2.VideoCapture(0)
if not cap.isOpened():
    cap=cv2.VidepCapture(1)
    raise IOError("Cannot open webcam")
while True:
    ret,frame=cap.read()
    result=dep.analyze(frame,actions=['emotion'],enforce_detection=False)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=faceCascade.detectMultiScale(gray,1.1,4)
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,result['dominant_emotion'],(0,50),font,2,(0,0,255),1,cv2.LINE_4);
    cv2.imshow("Orignal Video",frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()    
```
